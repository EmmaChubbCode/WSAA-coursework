```python
import requests
```

```python
# get the exchequer data from CSO
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/1.0/en"
response = requests.get(url)
```

```python
import json
```

```python
# turn whatever was returned into json so i can read it: https://developer.mozilla.org/en-US/docs/Web/API/Response/json 
data = response.json()
```

```python
#write the data to a file using dump, see: https://www.geeksforgeeks.org/python/read-json-file-using-python/ 
with open ('cso.json', 'w') as f:
    json.dump(data, f, indent=2)
```
